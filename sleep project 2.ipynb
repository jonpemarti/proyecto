```python
sleep_project2 = pd.read_csv('sleep_project2.csv')
sleep_project2
```

```python


# Definir el mapeo de edades a generaciones
generations_mapping = {
    (0, 25): 'Generacion Z',
    (26, 40): 'Millennials',
    (41, 56): 'Generacion X',
    (57, 75): 'Baby Boomers'
}

# Función para asignar la generación según la edad
def assign_generation(age):
    for age_range, generation in generations_mapping.items():
        if age_range[0] <= age <= age_range[1]:
            return generation
    return 'Otra'

# Aplicar la función a la columna 'age' para obtener la columna 'generation'
sleep_project2['generation'] = sleep_project2['age'].apply(assign_generation)

# Mostrar el DataFrame con la nueva columna 'generation'
print(sleep_project2)
```

```python
from scipy.stats import ttest_ind

# Extraer las muestras para el test
sample1 = sleep_project2['sleep_duration']
sample2 = sleep_project2['sleep_efficiency']

# Realizar el test de hipótesis
t_statistic, p_value = ttest_ind(sample1, sample2, equal_var=False)

# Imprimir los resultados
print("Estadístico t:", t_statistic)
print("Valor p:", p_value)

# Comparar el valor p con un nivel de significancia (alfa) predefinido
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula (Ho) en favor de la hipótesis alternativa (H1)")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula (Ho)")
```
