```python
sleep_project2 = pd.read_csv('sleep_project2.csv')
sleep_project2
```

```python


# Definir el mapeo de edades a generaciones
generations_mapping = {
    (0, 25): 'Generacion Z',
    (26, 40): 'Millennials',
    (41, 56): 'Generacion X',
    (57, 75): 'Baby Boomers'
}

# Función para asignar la generación según la edad
def assign_generation(age):
    for age_range, generation in generations_mapping.items():
        if age_range[0] <= age <= age_range[1]:
            return generation
    return 'Otra'

# Aplicar la función a la columna 'age' para obtener la columna 'generation'
sleep_project2['generation'] = sleep_project2['age'].apply(assign_generation)

# Mostrar el DataFrame con la nueva columna 'generation'
print(sleep_project2)
```

```python
from scipy.stats import ttest_ind

# Extraer las muestras para el test
sample1 = sleep_project2['sleep_duration']
sample2 = sleep_project2['sleep_efficiency']

# Realizar el test de hipótesis
t_statistic, p_value = ttest_ind(sample1, sample2, equal_var=False)

# Imprimir los resultados
print("Estadístico t:", t_statistic)
print("Valor p:", p_value)

# Comparar el valor p con un nivel de significancia (alfa) predefinido
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula (Ho) en favor de la hipótesis alternativa (H1)")
else:
    print("No hay suficiente evidencia para rechazar la hipótesis nula (Ho)")
```

```python
from scipy.stats import pearsonr, pointbiserialr, spearmanr

# Columnas numéricas
numerical_columns = ['sleep_efficiency', 'age', 'stress_level', 'heart_rate', 
                     'daily_steps', 'rem_sleep_percentage', 'deep_sleep_percentage', 
                     'light_sleep_percentage', 'awakenings']

# Lista para almacenar los resultados de la correlación
correlation_results_dur = []

# Calcular correlaciones con columnas numéricas
print("Correlaciones de sleep duration con columnas numéricas:")
for column in numerical_columns:
    correlation, p_value = pearsonr(sleep_project2['sleep_duration'], sleep_project2[column])
    correlation_results_dur.append({'Column': column, 'Correlation': correlation, 'P-value': p_value})

# Convertir la lista de resultados en un DataFrame
correlation_duration_numerical = pd.DataFrame(correlation_results_dur)

correlation_duration_numerical= correlation_duration_numerical.sort_values(by='Correlation', ascending=False)

# Mostrar el DataFrame de correlaciones
correlation_duration_numerical.round(4)
```

```python
from sklearn.preprocessing import LabelEncoder

# Crear una copia del DataFrame para evitar modificar el original
sleep_project2_encoded = sleep_project2.copy()

# Inicializar el codificador de etiquetas
label_encoder = LabelEncoder()

# Iterar sobre cada columna categórica y codificar las categorías
for column in categorical_columns:
    # Convertir la columna a tipo 'category' para asegurarnos de que sea tratada como categórica
    sleep_project2_encoded[column] = label_encoder.fit_transform(sleep_project2_encoded[column].astype('category'))

# Verificar el nuevo DataFrame con categorías codificadas
print(sleep_project2_encoded.head())
```

```python
import pandas as pd
from scipy.stats import pointbiserialr, spearmanr

# Columnas categóricas
categorical_columns = ['gender', 'occupation', 'bmi_category', 'blood_pressure', 
                       'caffeine_consumption', 'alcohol_consumption', 'smoking_status', 
                       'exercise_frequency', 'generation']

# Lista para almacenar los resultados de la correlación
correlation_results_categorical_dur = []

# Calcular correlaciones con columnas categóricas
print("\nCorrelaciones de sleep duration con columnas categóricas:")
for column in categorical_columns:
    if len(sleep_project2_encoded[column].unique()) == 2:  # Si la columna categórica tiene solo dos categorías
        correlation, p_value = pointbiserialr(sleep_project2_encoded[column], sleep_project2_encoded['sleep_duration'])
    else:
        correlation, p_value = spearmanr(sleep_project2_encoded[column], sleep_project2_encoded['sleep_duration'])
    correlation_results_categorical_dur.append({'Column': column, 'Correlation': correlation, 'P-value': p_value})

# Convertir la lista de resultados en un DataFrame
correlation_duration_categorical = pd.DataFrame(correlation_results_categorical_dur)

correlation_duration_categorical= correlation_duration_categorical.sort_values(by='Correlation', ascending=False)


# Mostrar el DataFrame de correlaciones
correlation_duration_categorical.round(4)
```
