```python
import pandas as pd
import numpy as np

# Supongamos que tienes un DataFrame llamado 's1' con las columnas 'sleep_duration' y 'stress_level'
# y deseas realizar la sustituci√≥n descrita.

# Crear una copia del DataFrame para evitar modificar el original
s1_copy = s1.copy()

# Crear un diccionario de mapeo entre valores √∫nicos de 'sleep_duration' y 'stress_level'
mapping_dict = {}

for index, row in s1_copy.iterrows():
    sleep_duration = row['sleep_duration']
    stress_level = row['stress_level']
    
    # Verificar si el valor de 'stress_level' no es NaN
    if not pd.isna(stress_level):
        if sleep_duration in mapping_dict:
            # Si el valor de 'sleep_duration' ya est√° en el diccionario, agregar 'stress_level' a la lista
            mapping_dict[sleep_duration].append(stress_level)
        else:
            # Si es la primera vez que se encuentra el valor de 'sleep_duration', crear una nueva lista
            mapping_dict[sleep_duration] = [stress_level]

# Funci√≥n para sustituir NaN en 'stress_level' basado en 'sleep_duration'
def replace_nan_stress_level(row):
    sleep_duration = row['sleep_duration']
    stress_level = row['stress_level']
    
    # Verificar si el valor de 'stress_level' es NaN
    if pd.isna(stress_level):
        # Si hay m√∫ltiples valores de 'stress_level' para este 'sleep_duration', elegir aleatoriamente uno
        if sleep_duration in mapping_dict:
            stress_level_options = mapping_dict[sleep_duration]
            stress_level = np.random.choice(stress_level_options)
        else:
            # Si no hay informaci√≥n previa, mantener NaN
            stress_level = np.nan
    
    return stress_level

# Aplicar la funci√≥n de sustituci√≥n a la columna 'stress_level'
s1_copy['stress_level'] = s1_copy.apply(replace_nan_stress_level, axis=1)

# Mostrar el DataFrame resultante
print(s1_copy)
```

```python
import pandas as pd
import numpy as np

s1_copy = s1.copy()


mapping_dict = {}

for index, row in s1_copy.iterrows():
    sleep_efficiency = row['sleep_efficiency']
    awakenings = row['awakenings']
    
    
    if not pd.isna(awakenings):
        if sleep_efficiency in mapping_dict:
            
            mapping_dict[sleep_efficiency].append(awakenings)
        else:
            
            mapping_dict[sleep_efficiency] = [awakenings]


def replace_nan_awakenings(row):
    sleep_efficiency = row['sleep_efficiency']
    awakenings = row['awakenings']
    
    
    if pd.isna(awakenings):
        
        if sleep_efficiency in mapping_dict:
            awakenings_options = mapping_dict[sleep_efficiency]
            awakenings = np.random.choice(awakenings_options)
        else:
            
            awakenings = np.nan
    
    return awakenings


s1_copy['awakenings'] = s1_copy.apply(replace_nan_awakenings, axis=1)

print(s1_copy)
```
