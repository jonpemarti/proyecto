```python
import pandas as pd
import numpy as np

# Supongamos que tienes un DataFrame llamado 's1' con las columnas 'bmi_category' y 'blood_pressure'
# y deseas realizar la sustituci√≥n descrita.

# Crear una copia del DataFrame para evitar modificar el original
s1_copy = s1.copy()

# Crear un diccionario de mapeo entre valores √∫nicos de 'bmi_category' y 'blood_pressure'
mapping_dict = {}

for index, row in s1_copy.iterrows():
    blood_pressure = row['blood_pressure']
    sleep_disorder = row['sleep_disorder']
    
    # Verificar si el valor de 'blood_pressure' no es NaN
    if not pd.isna(sleep_disorder):
        if blood_pressure in mapping_dict:
            # Si el valor de 'bmi_category' ya est√° en el diccionario, agregar 'blood_pressure' a la lista
            mapping_dict[blood_pressure].append(sleep_disorder)
        else:
            # Si es la primera vez que se encuentra el valor de 'bmi_category', crear una nueva lista
            mapping_dict[blood_pressure] = [sleep_disorder]

# Funci√≥n para sustituir NaN en 'blood_pressure' basado en 'bmi_category'
def replace_nan_sleep_disorder(row):
    blood_pressure = row['blood_pressure']
    sleep_disorder = row['sleep_disorder']
    
    # Verificar si el valor de 'blood_pressure' es NaN
    if pd.isna(sleep_disorder):
        # Si hay m√∫ltiples valores de 'blood_pressure' para este 'bmi_category', elegir aleatoriamente uno
        if blood_pressure in mapping_dict:
            sleep_disorder_options = mapping_dict[blood_pressure]
            sleep_disorder = np.random.choice(sleep_disorder_options)
        else:
            # Si no hay informaci√≥n previa, mantener NaN
            sleep_disorder = np.nan
    
    return sleep_disorder

# Aplicar la funci√≥n de sustituci√≥n a la columna 'blood_pressure'
s1_copy['sleep_disorder'] = s1_copy.apply(replace_nan_sleep_disorder, axis=1)

# Mostrar el DataFrame resultante
print(s1_copy)
```

```python
import pandas as pd
import numpy as np

# Supongamos que tienes un DataFrame llamado 'df'

# Obt√©n los tipos √∫nicos de valores en la columna sleep_disorder
valores_unicos = s1['sleep_disorder'].unique()

# Reemplaza los NaN por valores √∫nicos aleatorios
s1['sleep_disorder'] = s1['sleep_disorder'].apply(lambda x: np.random.choice(valores_unicos) if pd.isna(x) else x)

# Reemplaza cualquier NaN restante con un valor √∫nico aleatorio
valores_unicos = s1['sleep_disorder'].unique()
s1['sleep_disorder'].fillna(np.random.choice(valores_unicos), inplace=True)

# Mostrar el DataFrame resultante
print(s1)
```
